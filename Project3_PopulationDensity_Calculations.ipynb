```python
df_ger['percentage'] = df_ger['percentage']*100
```

```python
df_ger.to_csv("wholegermany.csv", index=False)
```

# Bind the Datasets together

```python
df = pd.concat([df_bw, df_by, df_ber, df_br, df_bre, df_hh, df_hes, df_meck, df_nrw, df_n, df_rp, df_saar, df_sac, df_sa, df_sh, df_t])
```

```python
df
```

```python
df_nopop = df[df['value']==0.0].sort_values(by="percentage")
df_nopop['value'] = "No Population"
df_nopop['percentage'] = df_nopop['percentage']*100
df_nopop
```

```python
df_rural = df[df['value']==1.0].sort_values(by="percentage")
df_rural['value'] = "Rural Area"
df_rural['percentage'] = df_rural['percentage']*100
df_rural
```

```python
df_urb = df[df['value']==2.0].sort_values(by="percentage")
df_urb['value'] = "Urban Clusters"
df_urb['percentage'] = df_urb['percentage']*100
df_urb
```

```python
df_center = df[df['value']==3.0].sort_values(by="percentage")
df_center['value'] = "Urban Center"
df_center['percentage'] = df_center['percentage']*100
df_center
```

```python
df_all = pd.concat([df_nopop, df_rural, df_urb, df_center])
df_all
```

```python
df_total_populated = df_all[df_all['value']!='No Population'].groupby('state')['m2'].sum().sort_values()
df_total_populated
```

```python
df_pivot = df_all.pivot(index='state', columns='value', values='percentage')
```

```python
df_pivot = df_pivot[['No Population', 'Rural Area','Urban Clusters', 'Urban Center']]
df_pivot
```

```python
df_pivot.to_csv("allstates.csv", index=True)
```
